```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
train_data = pd.read_csv('/Users/arslanaliawan/Desktop/scripts/house_prices/train.csv')
```

```python

train_data['SalePrice'] = np.log1p(train_data['SalePrice'])
plt.subplots(figsize=(12,9))
sns.distplot(train_data['SalePrice'], fit=stats.norm)

# Get the fitted parameters used by the function

(mu, sigma) = stats.norm.fit(train_data['SalePrice'])

# plot with the distribution

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)], loc='best')
plt.ylabel('Frequency')


fig = plt.figure()
stats.probplot(train_data['SalePrice'], plot=plt)
plt.show()
```
